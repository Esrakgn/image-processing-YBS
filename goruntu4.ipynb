```python
import os
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
my_binary_1_img = cv2.imread(r"C:\Users\Esra\Downloads\paint.png", 0)
```

```python
print(type(my_binary_1_img))
```

```python
my_kernel = np.ones((5,5), dtype=np.uint8)
```

my_erosion_img = cv2.erode(my_binary_1_img , my_kernel , iterations=6)

```python
#Erosion (aşındırma), görüntü işleme (morfolojik işlemler) içinde kullanılan bir yöntemdir.
#Özellikle binary (siyah-beyaz) görüntülerde uygulanır.
#Beyaz (255) alanları küçültür, siyah (0) alanları genişletir.
#Yani beyaz nesnelerin kenarları aşındırılır.
```

```python
fig = plt.figure(figsize=(10,7))

plt.subplot(1,2,1)
plt.imshow(my_binary_1_img,cmap = "gray")
plt.axis("off")
plt.title("Original Image")

plt.subplot(1,2,2)
plt.imshow(my_erosion_img,cmap = "gray")
plt.axis("off")
plt.title("Erosion Image")
```

```python
my_dilation_img = cv2.dilate(my_binary_1_img , my_kernel , iterations=6)
```

```python
fig = plt.figure(figsize=(10,7))
#

plt.subplot(2,2,1)
plt.imshow(my_binary_1_img,cmap = "gray")
plt.axis("off")
plt.title("Original Image")

plt.subplot(2,2,2)
plt.imshow(my_erosion_img,cmap = "gray")
plt.axis("off")
plt.title("Erosion Image")

plt.subplot(2,2,3)
plt.imshow(my_dilation_img,cmap = "gray")
plt.axis("off")
plt.title("Dilation Image")
```

```python
new_img =cv2.imread("./ybs.png",0)
```

```python
print(type(new_img))
```

```python
#beyaz renkli bir noise oluşturuyoruz(istenmeyen piksel)
my_white_noise = np.random.randint(0,2, size= new_img.shape[:2])
my_white_noise *=255
```

```python
plt.imshow(my_white_noise, cmap= "gray")
plt.axis("off")
plt.title("White Noise")
```

```python
my_noised_img_1 = my_white_noise + new_img
```

```python
fig = plt.figure(figsize=(10,7))

plt.subplot(1,2,1)
plt.imshow(new_img, cmap= "gray")
plt.axis("off")
plt.title("White Noise")

plt.subplot(1,2,2)
plt.imshow(my_noised_img_1,cmap = "gray")
plt.axis("off")
plt.title("Noised Image")
```

```python
my_opening_img =cv2.morphologyEx(my_noised_img_1.astype(np.float32), cv2.MORPH_OPEN,my_kernel)
```

```python
fig = plt.figure(figsize=(10,7))
plt.subplot(1,3,1)
plt.imshow(new_img, cmap= "gray")
plt.axis("off")
plt.title("White Noise")

plt.subplot(1,3,2)
plt.imshow(my_noised_img_1,cmap = "gray")
plt.axis("off")
plt.title("Noised Image")

plt.subplot(1,3,3)
plt.imshow(my_opening_img,cmap = "gray")
plt.axis("off")
plt.title("Opening Image")
```

```python
#siyah gürültü oluşturma 

my_black_noise =np.random.randint(0,2, size= new_img.shape[:2])
my_black_noise *= -255
```

```python
my_noised_img_2 = my_black_noise + new_img
```

```python
my_noised_img_2[my_noised_img_2<=-245] =0
```

```python
plt.imshow(my_noised_img_2,cmap ="gray")
plt.axis("off")
plt.title("Black Noise")
```

```python
my_closing_img =cv2.morphologyEx(my_noised_img_2.astype(np.float32), cv2.MORPH_CLOSE,my_kernel)
```

```python
fig = plt.figure(figsize=(10,7))
plt.subplot(1,3,1)
plt.imshow(new_img, cmap= "gray")
plt.axis("off")
plt.title("White Noise")

plt.subplot(1,3,2)
plt.imshow(my_noised_img_2,cmap = "gray")
plt.axis("off")
plt.title("Noised Image")

plt.subplot(1,3,3)
plt.imshow(my_closing_img,cmap = "gray")
plt.axis("off")
plt.title("Closing Image")
```

```python
my_gradient_1 = cv2.morphologyEx(my_binary_1_img, cv2.MORPH_GRADIENT, my_kernel)
my_gradient_2 = cv2.morphologyEx(new_img, cv2.MORPH_GRADIENT,my_kernel)
```

```python
fig = plt.figure(figsize=(10,7))
plt.subplot(2,2,1)
plt.imshow(my_binary_1_img, cmap= "gray")
plt.axis("off")
plt.title("Original Image")

plt.subplot(2,2,2)
plt.imshow(my_gradient_1, cmap= "gray")
plt.axis("off")
plt.title("Gradient 1 ")

plt.subplot(2,2,3)
plt.imshow(new_img,cmap = "gray")
plt.axis("off")
plt.title("New Image")

plt.subplot(2,2,4)
plt.imshow(my_gradient_2,cmap = "gray")
plt.axis("off")
plt.title("Gradient 2 Image")
```
