```python
import cv2

def img_zoom_in(img_org):
    new_height = int(img_org.shape[0] * 1.25)   # %25 büyüt
    new_width  = int(img_org.shape[1] * 1.25)
    new_img = cv2.resize(img_org, (new_width, new_height), interpolation=cv2.INTER_LANCZOS4)
    return new_img

def img_zoom_out(img_org):
    new_height = int(img_org.shape[0] * 0.75)   # %25 küçült
    new_width  = int(img_org.shape[1] * 0.75)
    new_img = cv2.resize(img_org, (new_width, new_height), interpolation=cv2.INTER_AREA)
    return new_img

```

```python
my_rgb_img =cv2.cvtColor(my_img, cv2.COLOR_BGR2RGB)
```

```python
new_big_img = img_zoom_in(my_rgb_img)
new_small_img = img_zoom_out(my_rgb_img)

print(f"Orijinal Görüntü = {my_rgb_img.shape[:2]}")
print(f"Büyütülmüş Görüntü = {new_big_img.shape[:2]}")
print(f"Küçültülmüs Görüntü = {new_small_img.shape[:2]}")
```
